Código desenvolvido para importação de dados MERGE-GPM diários do INPE: http://ftp.cptec.inpe.br/modelos/tempo/MERGE/GPM/DAILY/

Formato dos dados: "*.grib2"

```python
import os
from bs4 import BeautifulSoup
import requests
import csv

#Sequencia dos anos possiveis

for ano in range(2010,2023):
    anos = str(ano)
    links = 'http://ftp.cptec.inpe.br/modelos/tempo/MERGE/GPM/DAILY/'
    url = links + anos
    print(url)

    path = 'G:/Pos_Doutorado/MapbiomasAR/Precipitacao_MERGE/'
    caminho = path + anos

    isExist = os.path.exists(caminho)

    if not isExist:
  
        #Create a new directory because it does not exist 
        os.makedirs(caminho)
        print("The new directory is created!")
    
        os.chdir(caminho) 


    for k in range(1,13) :
    
        if k in range(10,13):
            ext = str(k)+"/"
            print(ext)
        else:
            #ext ; meses dentro da pasta de um ano
            a0 = "0"
            ext = a0+str(k)+"/"
            print(ext)
    
        dirr = caminho + "/" + ext
        print(dirr)
    
        isExist = os.path.exists(dirr)

        if not isExist:
  
            #Create a new directory because it does not exist 
            os.makedirs(dirr)
        print("The new directory is created!")
    
        os.chdir(dirr)        
            
    
        def listFD(url, ext=''):
            page = requests.get(url).text
            soup = BeautifulSoup(page, 'html.parser')
            return [url + '/' + node.get('href') for node in soup.find_all('a') if node.get('href').endswith(ext)]
    
        for file in listFD(url, ext):
       
            print(file)
    
            for file2 in listFD(file,'grib2'):
            #print(file2)
        
        
                url2 = file2
        
                r = requests.get(url2, allow_redirects=True)
                #print(str(file2))
                k = str(file2)
                print(k[64:90])
                # Define the local filename to save data
                local_file = k[64:90]
        
                with open(local_file, 'wb')as file:
                    file.write(r.content)                
    
```
